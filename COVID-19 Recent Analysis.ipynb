```python
plt.figure(figsize=(20,6))

plt.plot(df.date,df.test,color="blue",label="Test Says")
plt.plot(df.date,df.event,color="purple",label="Vaka Says") 
plt.plot(df.date,df.dead,color="red",label="lm Says")
plt.plot(df.date,df.heal,color="green",label="yileen Says")
plt.xticks(rotation=40)


plt.xlabel("Tarih")
plt.ylabel("Vaka ve lm Oran")
plt.title("COVID-19 Vaka & lm Oranlar")

plt.legend(loc=0)
plt.figure(figsize=(50,10))

plt.subplot(2,2,1)
plt.plot(df.date,df.event,color="blue",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Vaka Says")
plt.title("COVID-19 Grafii")
plt.xlabel('Tarih')
plt.ylabel('Vaka Says')
plt.xticks(rotation=40)
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()
```

```python
plt.figure(figsize=(50,10))
plt.subplot(2,2,1)
plt.plot(df.date,df.test,color="blue",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Test Says")
plt.title("COVID-19 Grafii")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('Test Says',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()

plt.figure(figsize=(50,10))
plt.subplot(2,2,2)
plt.plot(df.date,df.event,color="purple",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Vaka Says")
plt.title("COVID-19 Grafii")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('Vaka Says',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()


plt.figure(figsize=(50,10))
plt.subplot(2,2,3)
plt.plot(df.date,df.dead,color="red",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Vefat Says")
plt.title("COVID-19 Grafii")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('Vefat Eden Hasta Says',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()

plt.figure(figsize=(50,10))
plt.subplot(2,2,4)   
plt.plot(df.date,df.heal,color="green",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="yileen Says")
plt.title("COVID-19 Grafii")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('yileen Hasta Says',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik2.png",dpi=1200,quality=95)
plt.show()
```
