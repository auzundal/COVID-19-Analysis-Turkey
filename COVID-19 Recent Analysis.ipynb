```python
%matplotlib inline

import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')
```

```python
plt.figure(figsize=(20,6))

plt.plot(df.date,df.test,color="blue",label="Test Sayısı")
plt.plot(df.date,df.event,color="purple",label="Vaka Sayısı") 
plt.plot(df.date,df.dead,color="red",label="Ölüm Sayısı")
plt.plot(df.date,df.heal,color="green",label="İyileşen Sayısı")
plt.xticks(rotation=40)


plt.xlabel("Tarih")
plt.ylabel("Vaka ve Ölüm Oranı")
plt.title("COVID-19 Vaka & Ölüm Oranları")

plt.legend(loc=0)
plt.figure(figsize=(50,10))

plt.subplot(2,2,1)
plt.plot(df.date,df.event,color="blue",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Vaka Sayısı")
plt.title("COVID-19 Grafiği")
plt.xlabel('Tarih')
plt.ylabel('Vaka Sayısı')
plt.xticks(rotation=40)
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()
```

```python
plt.figure(figsize=(50,10))
plt.subplot(2,2,1)
plt.plot(df.date,df.test,color="blue",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Test Sayısı")
plt.title("COVID-19 Grafiği")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('Test Sayısı',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()

plt.figure(figsize=(50,10))
plt.subplot(2,2,2)
plt.plot(df.date,df.event,color="purple",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Vaka Sayısı")
plt.title("COVID-19 Grafiği")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('Vaka Sayısı',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()


plt.figure(figsize=(50,10))
plt.subplot(2,2,3)
plt.plot(df.date,df.dead,color="red",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="Vefat Sayısı")
plt.title("COVID-19 Grafiği")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('Vefat Eden Hasta Sayısı',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik1.png",dpi=1200,quality=95)
plt.show()

plt.figure(figsize=(50,10))
plt.subplot(2,2,4)   
plt.plot(df.date,df.heal,color="green",linewidth=2,marker="o",markersize=5,markerfacecolor="black",label="İyileşen Sayısı")
plt.title("COVID-19 Grafiği")
plt.xlabel('Tarih',fontsize=20, color='black')
plt.xticks(rotation=40)
plt.ylabel('İyileşen Hasta Sayısı',fontsize=20, color='black')
plt.legend(loc=0)
plt.grid(alpha=0.2)
#plt.savefig(fname="grafik2.png",dpi=1200,quality=95)
plt.show()
```
